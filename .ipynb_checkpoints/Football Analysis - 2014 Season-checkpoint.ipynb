***La Liga 2014 Analysis of the Goal Tally Distribution among the teams***

```python
Sym_liga = []
for i in range(20):
    j = 2
    team_short = la_liga_2014.team[i][0:3]
    if team_short in Sym_liga:
        team_short = team_short + ' -' + str(j)
        j+=1
    Sym_liga.append(team_short) 
```

```python
plt.title('La-Liga 2014 Goals Scored')

liga_plot_2014 = la_liga_2014.copy()
liga_plot_2014['teams_sym'] = Sym_liga

sns.barplot(x='teams_sym', y='scored', data=liga_plot_2014, color='#3AEFCA')
plt.xlabel('Teams')
plt.ylabel('Goals');
```

```python
la_liga_2014.scored.sum()
```

```python
liga_goal_data = []
liga_2014_top5 = liga_goal_data.append(liga_plot_2014[0:5].scored.sum())
liga_2014_rest15 = liga_goal_data.append(liga_plot_2014[5:20].scored.sum())
labels = ['La Liga 2014 Top 5', 'La Liga 2014 Rest 15']
plt.title('La Liga 2014 Goal Distribution : Top 5 vs Rest 15')
colors = ['#66b3ff','#99ff99','#ffcc99']

plt.pie(x=liga_goal_data, autopct="%.1f%%", explode=[0.05]*2, labels=labels, pctdistance=0.5, colors=colors);
```

#### Q4: Rank the attacking prowess of the leagues based on the avergae Expected Goals in the 2014 season.

```python
xg_avg_2014 = {}

xg_avg_2014['Serie-A'] = SerieA_2014.xG.mean()
xg_avg_2014['EPL'] = EPL_2014.xG.mean()
xg_avg_2014['La-Liga'] = la_liga_2014.xG.mean()
xg_avg_2014['Bundesliga'] = Bundesliga_2014.xG.mean()
```

```python
plt.plot(xg_avg_2014.keys(), xg_avg_2014.values(), 'o--c')

plt.ylim([44,51])
plt.xlabel('Leagues')
plt.ylabel('Average Expected Goals')
plt.title("Average Expected Goals Comparison : 2014 Season");
```

*The Spanish League had the highest average of Expected Goals throughout the 2014 season followed by the Italian League, the English Premier League and the German League. xG or the Expected Goal is the measure of the probability of a given shot resulting into a goal. Looking at a historical data set of lots of shots takes from lots of positions, and looking at which of those shots resulted in a goal, xG shows how many goals would an average team would be expected to score from those shots. Thus, it can lead us to the inference that xG is a somewhat better measure to compare the attacking prowess of teams in Football.*
